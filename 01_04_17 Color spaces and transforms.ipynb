```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('1_1_Image_Representation/images/water_balloons.jpg')

# Make a copy of the image
image_copy = np.copy(image)

# Change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
# RGB Channels

r = image_copy[:,:,0]
g = image_copy[:,:,1]
b = image_copy[:,:,2]

f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,10))

ax1.set_title('Red')
ax1.imshow(r,cmap='gray')

ax2.set_title('Green')
ax2.imshow(g,cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b,cmap='gray')
```

```python
# Converting from RGB to HSV -> Hue, Saturation, Value
hsv = cv2.cvtColor(image_copy,cv2.COLOR_RGB2HSV)

# HSV channels
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

f, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h,cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s,cmap='gray')

ax3.set_title('Value')
ax3.imshow(v,cmap='gray')

ax4.set_title('Original')
ax4.imshow(image_copy)
```

#### Define pink and hue selection thresholds

```python
# Define our color selection criteria in RGB values
lower_pink = np.array([180,0,100])
upper_pink = np.array([255,255,230])

# Define our color selection criteria in HSV values
lower_hue = np.array([160,0,0])
upper_hue = np.array([180,255,255])
```

#### Mask the image

```python
#RGB mask

# Define the masked area
mask_rgb = cv2.inRange(image_copy, lower_pink, upper_pink)

masked_image = np.copy(image_copy)
masked_image[mask_rgb == 0] = [0,0,0]

f, (ax1,ax2) = plt.subplots(1,2,figsize=(20,10))

ax1.set_title('Original')
ax1.imshow(image_copy)

ax2.set_title('After mask')
ax2.imshow(masked_image)
```

```python
# HSV mask

# Define the masked area
mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)

# Create an copy and mask the image
masked_image = np.copy(image_copy)
masked_image[mask_hsv == 0] = [0,0,0]

f, (ax1,ax2) = plt.subplots(1,2,figsize=(20,10))

ax1.set_title('Original')
ax1.imshow(image_copy)

ax2.set_title('After mask')
ax2.imshow(masked_image)
```
